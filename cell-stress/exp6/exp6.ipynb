```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
excel_file = r'C:\Users\yda2\Desktop/w.xlsx'
df = pd.read_excel(excel_file)
df
```

```python
csv_file = 'fiji-output.csv'
df.to_csv(csv_file, index=False)
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(12, 6), facecolor="lightgray")
ax = plt.axes()
ax.set_facecolor("whitesmoke")

# Set grid with major and minor ticks
ax.grid(which='major', linestyle=':', linewidth='0.3', color='gray')
ax.grid(which='minor', linestyle=':', linewidth='0.3', color='gray')
ax.minorticks_on()

# Plot the data
plt.plot(df["time [min]"], df["Cell 1"], label="Cell 1", linewidth=3, color="slategray")
plt.plot(df["time [min]"], df["Cell 2"], label="Cell 2", linewidth=3, color="slategray")
plt.plot(df["time [min]"], df["Cell 3"], label="Cell 3", linewidth=3, color="steelblue")
plt.plot(df["time [min]"], df["Cell 4"], label="Cell 4", linewidth=3, color="slategray")
plt.plot(df["time [min]"], df["Cell 5 "], label="Cell 5", linewidth=3, color="slategray")

# Set the limits and labels
plt.xlim(100, 600)
plt.ylabel("CFP/FRET (arbitrary units, u.a.)", fontsize=14, fontweight="bold")
plt.xlabel("Time (min)", fontsize=14, fontweight="bold")
plt.title("Apoptosis Time-Lapse", fontsize=16, fontweight="bold")
plt.legend()
plt.savefig('apa.png', dpi=300, bbox_inches='tight')
plt.show()
```
